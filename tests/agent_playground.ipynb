# Agent Interactive Playground

Use this notebook to interactively test and debug the Agent's Planner, Validator, and Executor components.

```python
# [Setup] Environment & Imports
import sys
import os
import logging
import json
import asyncio

# Ensure project root is in path
# Dynamically add project root to sys.path
current_dir = os.getcwd()
while True:
    if os.path.exists(os.path.join(current_dir, 'agent', '__init__.py')):
        if current_dir not in sys.path:
            sys.path.append(current_dir)
            print(f"Added {current_dir} to sys.path")
        break
    parent = os.path.dirname(current_dir)
    if parent == current_dir:
        # Fallback: Assume parent of CWD is root
        fallback = os.path.abspath(os.path.join(os.getcwd(), '..'))
        if fallback not in sys.path:
            sys.path.append(fallback)
            print(f"Added {fallback} to sys.path (fallback)")
        break
    current_dir = parent

# Configure Logging (Suppress noise)
logging.getLogger("sentence_transformers").setLevel(logging.WARNING)
logging.getLogger("urllib3").setLevel(logging.WARNING)
logging.getLogger("openai").setLevel(logging.WARNING)
logging.basicConfig(level=logging.INFO)

from agent.core.planner import Planner
from agent.core.validator import PlanValidator
from agent.core.executor import ToolExecutor
from agent.core.schema import ExecutionContext
from agent.tools.manager import ToolManager
from agent.memory.manager import MemoryManager
from agent.tools.base import BaseTool

print("Environment Ready.")
```

```python
# [Init] Initialize Components
memory = MemoryManager()
tool_manager = ToolManager(memory_manager=memory)
planner = Planner()
validator = PlanValidator()
executor = ToolExecutor(tool_manager)
context = ExecutionContext()

print("Components Initialized.")
```

```python
# [Define Tools] Add Custom/Test Tools here

# Example: Mock RAG Tools (if not loaded from skills/)
class MockListFiles(BaseTool):
    def __init__(self):
        super().__init__("list_files", "List files in directory")
    def run(self, directory: str):
        return ["data.txt", "config.json"] 

class MockReadFile(BaseTool):
    def __init__(self):
        super().__init__("read_file", "Read file content")
    def run(self, file_path: str):
        return "Mock file content..."

# Register manually if needed (ToolManager loads from skills/ by default)
# tool_manager.register_tool(MockListFiles())
# tool_manager.register_tool(MockReadFile())

print("Available Tools:")
for t in tool_manager.list_tools(limit=20):
    print(f"- {t['name']}")
```

```python
# [Plan] Generate Plan
objective = "Analyze the architecture of the RAG system in ./tests/mock_rag"

print(f"Planning for: {objective}...")
available_tools = tool_manager.list_tools(limit=10)
plan = planner.plan(objective, available_tools)

print(json.dumps(plan.model_dump(), indent=2))
```

```python
# [Validate] Check Plan Validity
validation_result = validator.validate(plan)

if validation_result.valid:
    print("✅ Plan is Valid.")
else:
    print("❌ Plan is Invalid!")
    for err in validation_result.errors:
        print(f"  - {err}")
```

```python
# [Execute] Run Step-by-Step
# Reset context for fresh run
context = ExecutionContext()
context.set("input", objective)

print("Starting Execution...")

for step in plan.steps:
    print(f"\n▶ Step {step.step_id}: {step.intent}")
    try:
        # Execute step (await if running in async notebook environment)
        # Note: In standard Jupyter, top-level await is supported.
        result = await executor.execute_step(step, context)
        
        status_icon = "✅" if result.status == "success" else "❌"
        print(f"{status_icon} Status: {result.status}")
        if result.status == "success":
            print(f"   Result: {str(result.result)[:100]}...")
        else:
            print(f"   Error: {result.error}")
            if step.fallback_strategy == "fail":
                print("   Stopping due to failure.")
                break
    except Exception as e:
        print(f"   Critical Exception: {e}")
        break
```

```python
# [Inspect] View Final Context
print("Final Context Variables:")
for k, v in context.variables.items():
    print(f"- {k}: {str(v)[:100]}")
```
